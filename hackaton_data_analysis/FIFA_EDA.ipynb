```python
# changing the datatpes from str to float and doing with only value and wages:
def value_norm(x):
    x = str(x).replace('€', '')
    
    if('M' in str(x)):
        x = str(x).replace('M', '')
        x = float(x) * 1000000
        
    elif('K' in str(x)):
        x = str(x).replace('K', '')
        x = float(x) * 1000
    elif('" "' in str(x)):
        x = str(x).replace('""','')
        x = float(x)
        
    return float(x)

df['Value'] = df['Value'].apply(value_norm)
df['Wage'] = df['Wage'].apply(value_norm)
```

```python
# converting height into flaot where we, take a nwe list and append in it and change tha datatype and change into inch
# and those columns are empty it change take as nan in it :-
Height_cm = []
for i in list(df['Height'].values):
    try:
        Height_cm.append((float(str(i)[0])*12.0 + float(str(i)[2:]))*2.54)
    except(ValueError):
        Height_cm.append(np.nan)
        
df['Height'] = Height_cm
```

```python
df['Height']
```

```python
# converting weight into flaot where we, take a nwe list and append in it and change tha datatype and take first three element and take down in a new list
# and those columns are empty it change take as nan in it :-
Weight_cm = []
for i in list(df['Weight'].values):
    try:
        Weight_cm.append((float(str(i)[:3])))
    except(ValueError):
        Weight_cm.append(np.nan)
        
df['Weight'] = Weight_cm
```
